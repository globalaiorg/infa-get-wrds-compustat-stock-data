```python
import wrds
db = wrds.Connection(wrds_username='geert')
db.create_pgpass_file()
```

```python
db.describe_table(library="comp", table="secd")
```

```python
import pandas as pd
startdate = pd.to_datetime("2017-12-31").date()
enddate = pd.to_datetime("2019-02-27").date()
ticker_list=['AAPL','MSFT']
parm={'tickers':tuple(ticker_list),'start_date':startdate}
```

```python
data1 = db.raw_sql('SELECT tic,datadate,prccd,cshtrd FROM comp.secd WHERE tic in %(tickers)s AND datadate > %(start_date)s', params=parm)
data1
```
